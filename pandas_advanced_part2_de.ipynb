```python
# 8 a)

import seaborn as sns
sns.set_style("whitegrid")


plt.figure(figsize=(10,10))
sns.scatterplot(data=nba[(nba["_iscopy"]==0)& ((nba["team_id"]=="LAL") & (nba["opp_id"]== "GSW"))], x="pts", y="opp_pts", hue="game_result")
```

```python

# 8 b)
sns.displot(
    nba[(nba["_iscopy"]==0)&(nba["year_id"]>2010) & ((nba["team_id"]=="GSW") | (nba["team_id"]== "BOS"))], x="pts", col="fran_id", row="game_result",
    binwidth=5, height=4, facet_kws=dict(margin_titles=True),
)
```

```python
#8 c)

plt.figure(figsize=(30,10))
ax = sns.boxplot(data=nba[(nba["_iscopy"]==0)&(nba["year_id"]==2015)], x="fran_id", y="pts")

#Die Zeile unterhalb k√∂nnen Sie aktivieren in dem Sie das #-Zeichen l√∂schen. Nun werden die einzelnen Werte als Punkte auf die Boxplots gelegt.
#ax = sns.swarmplot(data=nba[(nba["_iscopy"]==0)&(nba["year_id"]==2015)], x="fran_id", y="pts", color = "0.25")
```
